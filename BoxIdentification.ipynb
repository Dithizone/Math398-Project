## Identification of each box

This notebook generates a few graphs from each box along the sensor path 11, 74, 137, 200, 242, which seems to start at the edge of the box and end at the center where the source is.

The identities of the boxes proven below in the graphs are:
- Box2: Cobalt-60
- Box3: Cesium-137

The remaining two boxes &mdash; Box1 and Box4 &mdash; are tentatively labeled KUT background and Technetium-99, respectively, though these are less obviously distinguished.

```python
import pandas as pd

print("Libraries imported.")
```

```python
Box1 = pd.read_csv("Box1.csv", header=None, index_col=None)
Box2 = pd.read_csv("Box2.csv", header=None, index_col=None)
Box3 = pd.read_csv("Box3.csv", header=None, index_col=None)
Box4 = pd.read_csv("Box4.csv", header=None, index_col=None)

print("Dataframes created for each box.")
```

```python
Box1.plot(kind="line", y=10, figsize=(8, 6), logy=True, title="Box 1")
```

```python
Box1.plot(kind="line", y=73, figsize=(8, 6), logy=True, title="Box 1")
```

```python
Box1.plot(kind="line", y=136, figsize=(8, 6), logy=True, title="Box 1")
```

```python
Box1.plot(kind="line", y=199, figsize=(8, 6), logy=True, title="Box 1")
```

```python
Box1.plot(kind="line", y=241, figsize=(8, 6), logy=True, title="Box 1")
```

## Tentative Assessment

Box 1 (above) is the KUT background. The sharp peak close to zero decreases and the region between between 0 and 0.2 MeV bumps upward slightly, like in Box 4. Here, however, there is no peak near 0.137 MeV which begins to rise, suggesting this is the background.
