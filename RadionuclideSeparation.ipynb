## *Introduction*
If the 1764 columns represent individual sensors in a 21x21 grid for each of 4 radionuclide sources &mdash; technetium-99, cobalt-60, cesium-137, and KUT background &mdash; and the columns are ordered and grouped by each type, there are a few trends which should be seen when each row, rather than each column, is graphed.

If the first 441 columns are for one radionuclide, the next 441 for another, and so on, then the behavior of certain energy peaks should change abruptly in each group of 441 columns.

Furthermore, if the columns march along each row of the 21x21 grid, the middle of the 21-sensor row will be closer to the source than the edges and the peaks should increase in intensity in a sinusoidal pattern, repeating every 21 columns. When the sensors enter the steel boxes, this should appear like a notable dip (like a bite mark) out of the sine wave shape.

And so, the way the columns of our data correspond to the geometry should be determinable from the behavior of each row at each radionuclide's signature energy peak as it's graphed.

According to Luttman, the peaks attributed to each radionuclide are:
- 1.173 and 1.332 MeV for Cobalt-60
